```python
from matplotlib import pyplot as one_plot
import cv2
```

```python
photo = cv2.imread("897.jpg")
photo = cv2.cvtColor(photo , cv2.COLOR_BGR2RGB)

aio_model = cv2.CascadeClassifier("model.xml")
face = aio_model.detectMultiScale(photo)
```

```python
a = face[0][0]
b = face[0][1]
c = face[0][2]
d = face[0][3]
photo = cv2.rectangle(photo , (a , b) , (a+c , b+d) , (120,255,150) , 7)
```

```python
one_plot.imshow(photo)
```
