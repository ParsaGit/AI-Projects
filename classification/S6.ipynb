```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
from sklearn import linear_model
import sklearn.metrics as sm 
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt 
```

```python
from sklearn import datasets
```

```python
x = digits.data
y = digits.target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = linear_model.LogisticRegression()
```

```python
x.shape
```

```python
X_train.shape
```

```python
model.fit(X_train, Y_train)
```

```python
out_robot = model.predict(X_test)
```

```python
out_robot
```

```python
Y_test.shape
```

```python
Y_test
```

```python
error_number = Y_test - out_robot
```

```python
error_number
```

```python
is_realized = 0
not_realized = 0

for i in error_number:
    if(i == 0):
        is_realized += 1
        
    if(i != 0):
        not_realized += 1

print(is_realized , '\033[2;32;1mnumbers has been realized by the model\033[0;0m')
print(not_realized ,"\033[2;36;1mnumbers has\033[2;31;1m not\033[0;0m \033[2;36;1mbeen realized by the model\033[0;0m")
```

```python
is_realized_percentage = (is_realized / Y_test.shape[0]) * 100
is_realized_percentage
```
